```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,confusion_matrix
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
nltk.download('stopwords')
stop_words=set(stopwords.words('turkish'))
stemmer=PorterStemmer()

def temizle(metin):
  kelimeler=metin.lower().split()
  temiz_kelimeler=[stemmer.stem(k) for k in kelimeler if k not in stop_words]
  return " ".join(temiz_kelimeler)
```

```python
# Eğitim verilerini biraz güncelleyelim
texts = [
    "Bu film harikaydı, çok beğendim",
    "Müthiş bir deneyimdi, herkese öneririm",
    "Hiç beğenmedim, zaman kaybıydı",
    "Kötü bir servis, çok kötü", # 'Kötü' kelimesini burada netleştirdik
    "Yemekler lezzetliydi, servis iyiydi"
]
labels = [1, 1, 0, 0, 1]

# Modeli tekrar eğitip aynı cümleyi soralım:
# yeni_yorum = ["Kötü bir servis"]
```

```python
#Metinleri Sayısal Vektörlere Dönüştürme
temiz_texts=[temizle(metin) for metin in texts]
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(temiz_texts)
#Kelime Bilgisini Görüntüleme(Bag of Words)
print("Kelimeler : ",vectorizer.get_feature_names_out())
```

```python
#Modeli Oluşturma
model=MultinomialNB()
#Modeli Eğitme
model.fit(X,labels)
#Yeni Yorum ile Test Etme
yorum_test=["Servis gerçekten çok kötü"]
temiz_yorum_test = [temizle(yorum_test[0])]
yeni_vector = vectorizer.transform(temiz_yorum_test)

tahmin=model.predict(yeni_vector)
sonuc= "Olumlu" if tahmin[0]==1 else "Olumsuz"

print(f"Yorum: {yorum_test[0]}")
print(f"Tahmin: {sonuc}")
```

```python
egitim_tahmini=model.predict(X)
basari=accuracy_score(labels,egitim_tahmini)
print(f"Modelin Eğitim Başarısı : %{basari * 100}")
```

```python
matrix=confusion_matrix(labels,egitim_tahmini)
print("Doğru Tahminler : ",matrix)
```
