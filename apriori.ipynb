```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

TransactionEncoder'ın işlevi:

Sözlük Oluşturma: Tüm sepetleri tarayıp içindeki benzersiz (unique) ürünlerin bir listesini çıkarır.

Varlık-Yokluk Matrisi (One-Hot Encoding): Her bir sepeti, bu sözlükteki ürünlerin olup olmadığını kontrol ederek True (Var) veya False (Yok) değerlerine dönüştürür.

Böylece her satır bir sepeti, her sütun ise bir ürünü temsil eder hale gelir. Eğer tabloyu 0 ve 1lerden oluşan bir matris gibi düşünürsek, Apriori bu matris üzerinde sütunları birleştirip "Kaç tane 1 alt alta denk gelmiş?" diye bakarak hızlıca hesaplama yapabilir.

```python
datasets=[["Kan Şekeri","Kolesterol","Tansiyon"],
          ["Kan Şekeri","Tansiyon"],
          ["Kolesterol","Vitamin D"],
          ["Kan Şekeri","Kolesterol","Tansiyon","Vitamin D"],
          ["Kan Şekeri","Tansiyon"]]
```

```python
te=TransactionEncoder()
te_ary=te.fit(datasets).transform(datasets)
df=pd.DataFrame(te_ary,columns=te.columns_)
print(df)
```

```python
frequent_items=apriori(df,min_support=0.5,use_colnames=True)
print(frequent_items)
```

```python
# metric: Hangi metriğe göre kuralları süzelim? (Güven seçtik)
# min_threshold: Güven en az kaç olmalı?

rules=association_rules(frequent_items,metric="confidence",min_threshold=0.7)
print(rules[["antecedents","consequents","support","confidence","lift"]])
print(len(rules))
```

```python
# Noktaları kırmızı ve büyük (s=100) yaparak çiziyoruz
plt.scatter(rules['support'], rules['confidence'], color='red', s=100)

# Eksenleri 0 ile 1.1 arasına sabitleyerek noktaların dışarıda kalmasını önleyelim
plt.xlim(0, 1.1)
plt.ylim(0, 1.1)

plt.xlabel('Destek (Support)')
plt.ylabel('Güven (Confidence)')
plt.title('Basit Kontrol Grafiği')
plt.show()
```
